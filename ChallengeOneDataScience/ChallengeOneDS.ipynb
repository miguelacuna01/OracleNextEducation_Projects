## Ingresos totales tiendas

```python
ingresos_total_tienda1 = tienda1['Precio'].sum()
print(f'Cantidad de ingresos de tienda 1: $ {ingresos_total_tienda1}')
ingresos_total_tienda2 = tienda2['Precio'].sum()
print(f'Cantidad de ingresos de tienda 2: $ {ingresos_total_tienda2}')
ingresos_total_tienda3 = tienda3['Precio'].sum()
print(f'Cantidad de ingresos de tienda 3: $ {ingresos_total_tienda3}')
ingresos_total_tienda4 = tienda4['Precio'].sum()
print(f'Cantidad de ingresos de tienda 4: $ {ingresos_total_tienda4}')
```

```python

#Genero una diccionario recopilando la informacion de los ingresos de las cuatros tiendas para calcular el maximo

tiendas = ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4']
ingresos_tiendas = [ingresos_total_tienda1, ingresos_total_tienda2, ingresos_total_tienda3, ingresos_total_tienda4]

dic_ingresos = dict(zip(tiendas, ingresos_tiendas))
dic_ingresos_ordenado = dict(sorted(dic_ingresos.items(), key=lambda item: item[1], reverse=True))



for tienda, ingreso in dic_ingresos_ordenado.items():
    print(f'La tienda {tienda} tuvo un ingreso total de: $ {ingreso}')
    if ingreso == max(dic_ingresos_ordenado.values()):
      max_ingreso = ingreso
      max_tienda = tienda
    acum_ingresos = sum(dic_ingresos_ordenado.values())
    distribucion_ingreso = ingreso / acum_ingresos * 100
    print(f'La distribucion del ingreso de la tienda {tienda} es de: {distribucion_ingreso:.2f}%')

print('_________________________________')

print(f'La tienda {max_tienda} tuvo el mayor ingreso, un total de: $ {max_ingreso}')

```

```python
cantidad_productos_tienda1 = tienda1['Producto'].count()
cantidad_productos_tienda2 = tienda2['Producto'].count()
cantidad_productos_tienda3 = tienda3['Producto'].count()
cantidad_productos_tienda4 = tienda4['Producto'].count()

print(f'La cantidad de productos vendidos en la tienda 1 es: {cantidad_productos_tienda1}')
print(f'La cantidad de productos vendidos en la tienda 2 es: {cantidad_productos_tienda2}')
print(f'La cantidad de productos vendidos en la tienda 3 es: {cantidad_productos_tienda3}')
print(f'La cantidad de productos vendidos en la tienda 4 es: {cantidad_productos_tienda4}')

cantidad_productos_tiendas = [cantidad_productos_tienda1, cantidad_productos_tienda2, cantidad_productos_tienda3, cantidad_productos_tienda4]

ingresos_producto_promedio_tiendas = [ingresos_total_tienda1 / cantidad_productos_tienda1, ingresos_total_tienda2 / cantidad_productos_tienda2, ingresos_total_tienda3 / cantidad_productos_tienda3, ingresos_total_tienda4 / cantidad_productos_tienda4]

dic_ingresos_producto_promedio_tiendas = dict(zip(tiendas, ingresos_producto_promedio_tiendas))

distribucion_ingresos_tiendas = ingresos_tiendas/sum(ingresos_tiendas) * 100

dic_distribucion_ingresos_tiendas = dict(zip(tiendas, distribucion_ingresos_tiendas))

# Dataframe ingreso promedio por producto y distribucion ingresos para cada una de las tiendas

df_ingresos_distribucion_tiendas = pd.DataFrame({
    'Tiendas': list(dic_ingresos_producto_promedio_tiendas.keys()),
    'Ingreso promedio por producto ($)': list(dic_ingresos_producto_promedio_tiendas.values()),
    'Distribución de ingresos (%)': list(dic_distribucion_ingresos_tiendas.values())
})
```

```python
import matplotlib.pyplot as plt

def generar_grafico_ingresos_distribucion_tiendas(dataframe, columna_categorias, columna_ingreso_promedio, columna_distribucion_ingresos):
    """
    Genera un gráfico combinado de barras y líneas para comparar el porcentaje de ventas y el porcentaje de ingresos.

    Parámetros:
        dataframe (pd.DataFrame): El DataFrame con los datos procesados.
        columna_categorias (str): La columna que contiene las categorías (eje X).
        columna_ingreso_promedio (str): La columna con los porcentajes de ventas (barras).
        columna_distribucion_ingresos (str): La columna con los porcentajes de ingresos (línea).

    Retorna:
        None: Muestra el gráfico.
    """
    # Extraer los datos necesarios
    categorias = dataframe[columna_categorias]
    porcentajes = dataframe[columna_ingreso_promedio]
    porcentajes_ingresos = dataframe[columna_distribucion_ingresos]

    # Crear la figura y los ejes
    fig, ax1 = plt.subplots(figsize=(12, 6))  # Ajusta el tamaño de la figura si es necesario

    # Barra para el porcentaje de ventas (eje izquierdo)
    ax1.bar(categorias, porcentajes, color='blue', label='Ingreso promedio por producto ($)')
    ax1.set_ylabel('Ingreso promedio por producto ($)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Crear un segundo eje para el porcentaje de ingresos (eje derecho)
    ax2 = ax1.twinx()  # Comparte el eje x
    ax2.plot(categorias, porcentajes_ingresos, color='red', marker='o', label='Distribución de ingresos (%)')
    ax2.set_ylabel('Distribución de ingresos (%)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')


    # Configurar el gráfico
    plt.title('Ingreso promedio ($) vs. Distribución de ingresos (%)')
    ax1.set_xlabel('Tienda')  # Etiqueta del eje X
    ax1.set_xticks(range(len(categorias)))  # Asegurarse de que las etiquetas coincidan con las categorías

    plt.tight_layout()  # Ajusta el diseño para evitar superposiciones

    # Mostrar el gráfico
    plt.show()
```

```python
# Llamar a la función para generar el gráfico
generar_grafico_ingresos_distribucion_tiendas(
    dataframe=df_ingresos_distribucion_tiendas,
    columna_categorias='Tiendas',
    columna_ingreso_promedio='Ingreso promedio por producto ($)',
    columna_distribucion_ingresos='Distribución de ingresos (%)'
)
```

## Distribucion de ingresos por año para cada tienda

```python
# Convert 'Fecha de Compra' to datetime in all DataFrames
tienda1['Fecha de Compra'] = pd.to_datetime(tienda1['Fecha de Compra'], format='%d/%m/%Y')
tienda2['Fecha de Compra'] = pd.to_datetime(tienda2['Fecha de Compra'], format='%d/%m/%Y')
tienda3['Fecha de Compra'] = pd.to_datetime(tienda3['Fecha de Compra'], format='%d/%m/%Y')
tienda4['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'], format='%d/%m/%Y')

# Now you can use .dt accessor
ingresos_tienda1_ano = tienda1.groupby(tienda1['Fecha de Compra'].dt.year)['Precio'].sum()
ingresos_tienda2_ano = tienda2.groupby(tienda2['Fecha de Compra'].dt.year)['Precio'].sum()
ingresos_tienda3_ano = tienda3.groupby(tienda3['Fecha de Compra'].dt.year)['Precio'].sum()
ingresos_tienda4_ano = tienda4.groupby(tienda4['Fecha de Compra'].dt.year)['Precio'].sum()

utilidad_tienda1_ano = ingresos_tienda1_ano - tienda1.groupby(tienda1['Fecha de Compra'].dt.year)['Costo de envío'].sum()
utilidad_tienda2_ano = ingresos_tienda2_ano - tienda2.groupby(tienda2['Fecha de Compra'].dt.year)['Costo de envío'].sum()
utilidad_tienda3_ano = ingresos_tienda3_ano - tienda3.groupby(tienda3['Fecha de Compra'].dt.year)['Costo de envío'].sum()
utilidad_tienda4_ano = ingresos_tienda4_ano - tienda4.groupby(tienda4['Fecha de Compra'].dt.year)['Costo de envío'].sum()

calificacion_tienda1_ano = tienda1.groupby(tienda1['Fecha de Compra'].dt.year)['Calificación'].mean()
calificacion_tienda2_ano = tienda2.groupby(tienda2['Fecha de Compra'].dt.year)['Calificación'].mean()
calificacion_tienda3_ano = tienda3.groupby(tienda3['Fecha de Compra'].dt.year)['Calificación'].mean()
calificacion_tienda4_ano = tienda4.groupby(tienda4['Fecha de Compra'].dt.year)['Calificación'].mean()


# Convert each Series to a DataFrame and add a 'Tienda' column
df1 = ingresos_tienda1_ano.reset_index().rename(columns={'Fecha de Compra': 'Año de Compra', 'Precio': 'Ingresos'})
df1['Tienda'] = 'Tienda 1'
df1['Utilidad'] = utilidad_tienda1_ano.values # Add 'Utilidad' to df1
df1['Calificacion'] = calificacion_tienda1_ano.values # Add 'Calificacion' to df1

df2 = ingresos_tienda2_ano.reset_index().rename(columns={'Fecha de Compra': 'Año de Compra', 'Precio': 'Ingresos'})
df2['Tienda'] = 'Tienda 2'
df2['Utilidad'] = utilidad_tienda2_ano.values # Add 'Utilidad' to df2
df2['Calificacion'] = calificacion_tienda2_ano.values # Add 'Calificacion' to df2

df3 = ingresos_tienda3_ano.reset_index().rename(columns={'Fecha de Compra': 'Año de Compra', 'Precio': 'Ingresos'})
df3['Tienda'] = 'Tienda 3'
df3['Utilidad'] = utilidad_tienda3_ano.values # Add 'Utilidad' to df3
df3['Calificacion'] = calificacion_tienda3_ano.values # Add 'Calificacion' to df3

df4 = ingresos_tienda4_ano.reset_index().rename(columns={'Fecha de Compra': 'Año de Compra', 'Precio': 'Ingresos'})
df4['Tienda'] = 'Tienda 4'
df4['Utilidad'] = utilidad_tienda4_ano.values # Add 'Utilidad' to df4
df4['Calificacion'] = calificacion_tienda4_ano.values # Add 'Calificacion' to df4


# Concatenate the dataframes
ingresos_tiendas_ano = pd.concat([df1, df2, df3, df4], ignore_index=True) # Remove df5-df8

# Reorder columns
ingresos_tiendas_ano = ingresos_tiendas_ano[['Año de Compra', 'Tienda', 'Ingresos', 'Utilidad', 'Calificacion']]

# Calculate total income for each year
total_ingresos_ano = ingresos_tiendas_ano.groupby('Año de Compra')['Ingresos'].transform('sum')
total_utilidad_ano = ingresos_tiendas_ano.groupby('Año de Compra')['Utilidad'].transform('sum')
# print(total_ingresos_ano)

# Calculate percentage of income for each store
ingresos_tiendas_ano['Porcentaje de Ingresos'] = (ingresos_tiendas_ano['Ingresos'] / total_ingresos_ano) * 100
ingresos_tiendas_ano['Porcentaje de utilidad'] = (ingresos_tiendas_ano['Utilidad'] / total_utilidad_ano) * 100

ingresos_tiendas_ano
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepara los datos para el gráfico
años = ingresos_tiendas_ano['Año de Compra'].unique()
tiendas = ingresos_tiendas_ano['Tienda'].unique()

# Define colores para cada tienda
colores_tiendas = {
    'Tienda 1': 'blue',
    'Tienda 2': 'red',
    'Tienda 3': 'green',
    'Tienda 4': 'orange'
}

# Crea la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Define el ancho de las barras
ancho_barra = 0.2

# Define las posiciones de las barras para cada año
posiciones_años = np.arange(len(años))

# Itera sobre las tiendas y crea las barras para cada año
for i, tienda in enumerate(tiendas):
    # Obtén los datos para la tienda actual
    datos_tienda = ingresos_tiendas_ano[ingresos_tiendas_ano['Tienda'] == tienda]

    # Crea las barras para cada año con el color correspondiente
    for j, año in enumerate(años):
        porcentaje = datos_tienda[datos_tienda['Año de Compra'] == año]['Porcentaje de Ingresos'].values[0]
        ax.bar(posiciones_años[j] + i * ancho_barra, porcentaje, ancho_barra,
               label=tienda if j == 0 else "",  # Etiqueta solo en la primera barra de cada tienda
               color=colores_tiendas[tienda])

# Configura las etiquetas del eje x
ax.set_xticks(posiciones_años + ancho_barra * (len(tiendas) - 1) / 2)
ax.set_xticklabels(años)

# Configura las etiquetas de los ejes
ax.set_ylabel('Porcentaje de Ingresos')
ax.set_xlabel('Año')
ax.set_title('Porcentaje de Ingresos por Año y Tienda')

# Agrega una leyenda (opcional, puedes comentar esta línea si no la necesitas)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Ajusta el diseño
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

# 2. Ventas por categoría

Se calcula la cantidad de productos vendidos por categoría en cada tienda. El objetivo es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

## Tienda 1: Cantidad de ventas segun la categoria

```python
def analizar_categorias(dataframe, columna_agrupacion, columnas_agg):
    """
    Analiza las categorías de un DataFrame agrupando por una columna específica y aplicando funciones de agregación.

    Parámetros:
        dataframe (pd.DataFrame): El DataFrame de entrada.
        columna_agrupacion (str): La columna por la cual se agruparán los datos.
        columnas_agg (dict): Diccionario con las columnas y funciones de agregación a aplicar.
                             Ejemplo: {'Categoría del Producto': 'count', 'Precio': 'sum'}

    Retorna:
        pd.DataFrame: Un DataFrame con los resultados del análisis, incluyendo porcentajes y columnas renombradas.
    """
    # Agrupar y aplicar funciones de agregación
    resultado = dataframe.groupby(columna_agrupacion).agg(**columnas_agg)

    # Ordenar por la cantidad de productos en orden descendente
    resultado = resultado.sort_values(by=list(columnas_agg.keys())[0], ascending=False)

    # Reiniciar el índice
    resultado = resultado.reset_index()

    # Calcular el porcentaje de productos
    primera_columna = list(columnas_agg.keys())[0]  # Primera columna de agregación
    resultado['Porcentaje'] = (resultado[primera_columna] / resultado[primera_columna].sum()) * 100
    resultado['Porcentaje'] = resultado['Porcentaje'].round(2)

    # Calcular el porcentaje de ingresos
    segunda_columna = list(columnas_agg.keys())[1]  # Segunda columna de agregación
    resultado['Porcentaje Ingresos'] = (resultado[segunda_columna] / resultado[segunda_columna].sum()) * 100
    resultado['Porcentaje Ingresos'] = resultado['Porcentaje Ingresos'].round(2)

    # Renombrar las columnas
    nuevos_nombres = {
        primera_columna: 'Cantidad de Productos',
        segunda_columna: 'Ingresos Totales'
    }
    resultado = resultado.rename(columns=nuevos_nombres)

    # Reordenar las columnas
    nuevo_orden = [
        columna_agrupacion, 'Cantidad de Productos', 'Porcentaje', 'Ingresos Totales', 'Porcentaje Ingresos'
    ]
    resultado = resultado[nuevo_orden]

    # Imprimir resultados generales
    print(f'Cantidad total de ventas: {resultado["Cantidad de Productos"].sum()}')
    print(f'Cantidad total de ingresos: {resultado["Ingresos Totales"].sum()}')
    print('Tabla: Categoría más vendidas')

    return resultado
```

```python
# Definir las columnas de agregación
columnas_agg = {
    'cantidad_productos': ('Categoría del Producto', 'count'),
    'ingresos_productos': ('Precio', 'sum')
}

# Llamar a la función
resultado1 = analizar_categorias(tienda1, 'Categoría del Producto', columnas_agg)

# Mostrar el resultado
resultado1
```

```python
import matplotlib.pyplot as plt

def generar_grafico(dataframe, columna_categorias, columna_porcentaje, columna_porcentaje_ingresos):
    """
    Genera un gráfico combinado de barras y líneas para comparar el porcentaje de ventas y el porcentaje de ingresos.

    Parámetros:
        dataframe (pd.DataFrame): El DataFrame con los datos procesados.
        columna_categorias (str): La columna que contiene las categorías (eje X).
        columna_porcentaje (str): La columna con los porcentajes de ventas (barras).
        columna_porcentaje_ingresos (str): La columna con los porcentajes de ingresos (línea).

    Retorna:
        None: Muestra el gráfico.
    """
    # Extraer los datos necesarios
    categorias = dataframe[columna_categorias]
    porcentajes = dataframe[columna_porcentaje]
    porcentajes_ingresos = dataframe[columna_porcentaje_ingresos]

    # Crear la figura y los ejes
    fig, ax1 = plt.subplots(figsize=(12, 6))  # Ajusta el tamaño de la figura si es necesario

    # Barra para el porcentaje de ventas (eje izquierdo)
    ax1.bar(categorias, porcentajes, color='blue', label='Porcentaje de Ventas')
    ax1.set_ylabel('Porcentaje de Ventas', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Crear un segundo eje para el porcentaje de ingresos (eje derecho)
    ax2 = ax1.twinx()  # Comparte el eje x
    ax2.plot(categorias, porcentajes_ingresos, color='red', marker='o', label='Porcentaje de Ingresos')
    ax2.set_ylabel('Porcentaje de Ingresos', color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    # Sincronizar las escalas de ambos ejes
    max_valor = max(max(porcentajes), max(porcentajes_ingresos))  # Encuentra el valor máximo entre ambos conjuntos de datos
    ax1.set_ylim(0, max_valor + 5)  # Establece el mismo límite superior para el eje izquierdo
    ax2.set_ylim(0, max_valor + 5)  # Establece el mismo límite superior para el eje derecho

    # Configurar el gráfico
    plt.title('Porcentaje de Ventas vs. Porcentaje de Ingresos por Categoría')
    ax1.set_xlabel('Categoría del Producto')  # Etiqueta del eje X
    ax1.set_xticks(range(len(categorias)))  # Asegurarse de que las etiquetas coincidan con las categorías
    ax1.set_xticklabels(categorias, rotation=60, ha='right')  # Rotar las etiquetas del eje X

    plt.tight_layout()  # Ajusta el diseño para evitar superposiciones

    # Mostrar el gráfico
    plt.show()
```

```python

# Llamar a la función para generar el gráfico
generar_grafico(
    dataframe=resultado1,
    columna_categorias='Categoría del Producto',
    columna_porcentaje='Porcentaje',
    columna_porcentaje_ingresos='Porcentaje Ingresos'
)
```

```python
def Productos_mas_menos_vendidos(dataframe):
  producto_mas_vendido = dataframe['Producto'].value_counts().idxmax()
  categoria_producto_mas_vendido = dataframe[dataframe['Producto'] == producto_mas_vendido]['Categoría del Producto'].iloc[0]
  producto_menos_vendido = dataframe['Producto'].value_counts().idxmin()
  categoria_producto_menos_vendido = dataframe[dataframe['Producto'] == producto_menos_vendido]['Categoría del Producto'].iloc[0]

  cantidad_mas_vendida = dataframe['Producto'].value_counts().max()
  cantidad_menos_vendida = dataframe['Producto'].value_counts().min()
  return producto_mas_vendido, categoria_producto_mas_vendido, cantidad_mas_vendida, producto_menos_vendido, categoria_producto_menos_vendido, cantidad_menos_vendida
```

## Utilidades por categoria para cada tienda

Crear un grafico de pareto

```python
def utilidad_por_categoria(dataframe):
  utilidad_categoria_tienda= (dataframe.groupby('Categoría del Producto')['Precio'].sum())-(dataframe.groupby('Categoría del Producto')['Costo de envío'].sum())
  utilidad_categoria_tienda = utilidad_categoria_tienda.sort_values(ascending=False)
  utilidad_categoria_tienda = utilidad_categoria_tienda.reset_index()
  utilidad_categoria_tienda = utilidad_categoria_tienda.rename(columns={'Categoría del Producto': 'Categoría del Producto', 0: 'Utilidad'})
  utilidad_categoria_tienda['Porcentaje'] = ((utilidad_categoria_tienda['Utilidad'] / utilidad_categoria_tienda['Utilidad'].sum()) * 100).round(2)
  utilidad_categoria_tienda['Porcentaje acumulado'] = utilidad_categoria_tienda['Porcentaje'].cumsum()

  # Mostrar el DataFrame con las columnas renombradas
  #utilidad_categoria_tienda1utilidad_categoria_tienda
  return utilidad_categoria_tienda
```

```python
# Grafico de pareto


def generar_grafico_pareto(dataframe, columna_categorias, col_utilidad, columna_utilidad_acum, tienda):
    """
    Genera un gráfico combinado de barras y líneas para comparar el porcentaje de ventas y el porcentaje de ingresos.

    Parámetros:
        dataframe (pd.DataFrame): El DataFrame con los datos procesados.
        columna_categorias (str): La columna que contiene las categorías (eje X).
        col_utilidad (str): La columna con los utilidad (barras).
        columna_utilidad_acum (str): La columna con los porcentajes de ingresos (línea).

    Retorna:
        None: Muestra el gráfico.
    """
    # Extraer los datos necesarios
    categorias = dataframe[columna_categorias]
    df_utilidad = dataframe[col_utilidad]
    porcentajes_acum = dataframe[columna_utilidad_acum]

    # Crear la figura y los ejes
    fig, ax1 = plt.subplots(figsize=(12, 6))  # Ajusta el tamaño de la figura si es necesario

    # Barra para el porcentaje de ventas (eje izquierdo)
    ax1.bar(categorias, df_utilidad, color='blue', label='Utilidad por categoria ($)')
    ax1.set_ylabel('Utilidad por categoria ($)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Crear un segundo eje para el porcentaje de ingresos (eje derecho)
    ax2 = ax1.twinx()  # Comparte el eje x
    ax2.plot(categorias, porcentajes_acum, color='red', marker='o', label='Porcentaje acumulado (%)')
    ax2.set_ylabel('Porcentaje acumulado (%)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')


    # Configurar el gráfico
    plt.title(f'Utilidad por categoria ($) vs. Porcentaje acumulado (%) {tienda}')
    ax1.set_xlabel('Utilidad por categoria')  # Etiqueta del eje X
    ax1.set_xticks(range(len(categorias)))  # Asegurarse de que las etiquetas coincidan con las categorías
    ax1.set_xticklabels(categorias, rotation=60, ha='right')  # Rotar las etiquetas del eje X

    plt.tight_layout()  # Ajusta el diseño para evitar superposiciones

    # Mostrar el gráfico
    plt.show()
```

```python
utilidad_categoria_tienda1 = utilidad_por_categoria(tienda1)
# Llamar a la función para generar el gráfico
generar_grafico_pareto(
    dataframe=utilidad_categoria_tienda1,
    columna_categorias='Categoría del Producto',
    col_utilidad='Utilidad',
    columna_utilidad_acum='Porcentaje acumulado',
    tienda='Tienda 1'
)

utilidad_categoria_tienda1
```

```python
utilidad_categoria_tienda3 = utilidad_por_categoria(tienda3)
# Llamar a la función para generar el gráfico
generar_grafico_pareto(
    dataframe=utilidad_categoria_tienda3,
    columna_categorias='Categoría del Producto',
    col_utilidad='Utilidad',
    columna_utilidad_acum='Porcentaje acumulado',
    tienda='Tienda 3'
)

utilidad_categoria_tienda3
```
